# Text-to-SQL V3 - Aggressive Training

**Strategy:** Higher learning rate, faster convergence

**Differences from V2:**
- Learning rate: 3e-4 (V2: 1e-4) - 3x more aggressive
- Epochs: 15 (V2: 20) - fewer but faster
- LR Schedule: linear (V2: cosine)
- Effective batch: 64 (V2: 32)
- Warmup: 10% (V2: 5%)
- Different prompt format

**Target:** 30-45% exact match

```python
# Cell 2: Imports and Setup
import torch
import numpy as np
import pandas as pd
import json
import re
import warnings
from collections import defaultdict

warnings.filterwarnings('ignore')

print("=" * 60)
print("TEXT-TO-SQL V3 - AGGRESSIVE")
print("=" * 60)
print(f"PyTorch: {torch.__version__}")
print(f"CUDA: {torch.cuda.is_available()}")

if torch.cuda.is_available():
    GPU_NAME = torch.cuda.get_device_name(0)
    GPU_MEM = torch.cuda.get_device_properties(0).total_memory / 1e9
    print(f"GPU: {GPU_NAME} ({GPU_MEM:.1f} GB)")
    MODEL_NAME = "google-t5/t5-base"
else:
    MODEL_NAME = "google-t5/t5-small"
    print("WARNING: No GPU!")

print(f"\nModel: {MODEL_NAME}")
print("=" * 60)
```

```python
# Cell 3: Load Spider Dataset
from datasets import load_dataset

print("Loading Spider dataset...")
dataset = None

for source in ["xlangai/spider", "spider"]:
    try:
        dataset = load_dataset(source)
        print(f"Loaded from {source}")
        break
    except:
        continue

if dataset is None:
    raise Exception("Could not load Spider dataset")

print(f"Train: {len(dataset['train'])} | Validation: {len(dataset['validation'])}")
```

```python
# Cell 4: Schema Serialization (V3 format - more explicit)
def serialize_schema_v3(example):
    """
    V3 format: TABLE table_name ( col1 TYPE, col2 TYPE, ... )
    More explicit SQL-like format.
    """
    try:
        table_names = example.get('db_table_names', [])
        column_names = example.get('db_column_names', [])
        column_types = example.get('db_column_types', [])
        
        table_cols = defaultdict(list)
        
        for idx, col_info in enumerate(column_names):
            if not isinstance(col_info, (list, tuple)) or len(col_info) < 2:
                continue
            
            table_idx, col_name = col_info[0], col_info[1]
            
            if table_idx == -1 or table_idx >= len(table_names):
                continue
            
            table_name = table_names[table_idx].lower()
            col_type = column_types[idx] if idx < len(column_types) else "text"
            col_str = f"{col_name} {col_type}"
            table_cols[table_name].append(col_str)
        
        if table_cols:
            parts = [f"TABLE {tbl} ( {', '.join(cols)} )" for tbl, cols in table_cols.items()]
            return " ; ".join(parts)
        return f"DATABASE {example.get('db_id', 'db')}"
    except:
        return f"DATABASE {example.get('db_id', 'db')}"

print("Schema function (V3 format) defined.")
```

```python
# Cell 5: SQL Normalization
def normalize_sql(sql):
    """
    Aggressive SQL normalization for better matching.
    """
    sql = sql.strip().lower()
    sql = re.sub(r'\s+', ' ', sql)
    sql = re.sub(r'\s*([=<>!]+)\s*', r' \\1 ', sql)
    sql = re.sub(r'\s*,\s*', ', ', sql)
    sql = re.sub(r'\s*\(\s*', '(', sql)
    sql = re.sub(r'\s*\)\s*', ') ', sql)
    sql = sql.rstrip(';').strip()
    sql = re.sub(r'\s+', ' ', sql).strip()
    return sql

print("SQL normalization ready.")
```

```python
# Cell 6: Preprocessing (V3 prompt format)
def preprocess_v3(example):
    question = str(example.get('question', '')).strip()
    sql = str(example.get('query', '')).strip()
    schema = serialize_schema_v3(example)
    
    # V3 format: cleaner, more explicit
    input_text = f"Question: {question} Schema: {schema} SQL:"
    target_text = normalize_sql(sql)
    
    return {
        "input_text": input_text,
        "target_text": target_text
    }

print("Preprocessing (V3 format)...")
processed = dataset.map(preprocess_v3, num_proc=4)

print(f"\nSample:")
print(f"Input: {processed['train'][0]['input_text'][:120]}...")
print(f"Target: {processed['train'][0]['target_text']}")
```

```python
# Cell 7: Tokenization
from transformers import AutoTokenizer

print(f"Loading tokenizer: {MODEL_NAME}")
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

MAX_INPUT = 512
MAX_TARGET = 256

def tokenize(examples):
    inputs = tokenizer(
        examples["input_text"],
        max_length=MAX_INPUT,
        truncation=True,
        padding=False
    )
    targets = tokenizer(
        text_target=examples["target_text"],
        max_length=MAX_TARGET,
        truncation=True,
        padding=False
    )
    inputs["labels"] = targets["input_ids"]
    return inputs

print("Tokenizing...")
tokenized = processed.map(
    tokenize,
    batched=True,
    num_proc=4,
    remove_columns=processed['train'].column_names
)

print(f"Done. Columns: {tokenized['train'].column_names}")
```

```python
# Cell 8: Load Model
from transformers import AutoModelForSeq2SeqLM

print(f"Loading model: {MODEL_NAME}")
model = AutoModelForSeq2SeqLM.from_pretrained(MODEL_NAME)
model.gradient_checkpointing_enable()

print(f"Parameters: {model.num_parameters():,}")
```

```python
# Cell 9: V3 AGGRESSIVE Training Config
from transformers import (
    Seq2SeqTrainingArguments,
    Seq2SeqTrainer,
    DataCollatorForSeq2Seq
)

data_collator = DataCollatorForSeq2Seq(
    tokenizer=tokenizer,
    model=model,
    label_pad_token_id=-100,
    padding=True
)

# V3 AGGRESSIVE SETTINGS
training_args = Seq2SeqTrainingArguments(
    output_dir="./t2sql_v3",
    
    # ===== V3 AGGRESSIVE SETTINGS =====
    num_train_epochs=15,           # Fewer epochs (V2: 20)
    learning_rate=3e-4,            # 3x higher (V2: 1e-4)
    warmup_ratio=0.1,              # More warmup (V2: 0.05)
    lr_scheduler_type="linear",    # Linear decay (V2: cosine)
    # ==================================
    
    per_device_train_batch_size=8,
    per_device_eval_batch_size=16,
    gradient_accumulation_steps=8,  # Larger effective batch: 64 (V2: 32)
    
    weight_decay=0.01,
    
    fp16=torch.cuda.is_available(),
    gradient_checkpointing=True,
    label_smoothing_factor=0.1,
    
    eval_strategy="steps",
    eval_steps=150,                 # More frequent eval (V2: 200)
    save_strategy="steps",
    save_steps=150,
    save_total_limit=3,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    
    predict_with_generate=True,
    generation_max_length=MAX_TARGET,
    generation_num_beams=4,
    
    logging_steps=50,
    report_to="none",
    dataloader_num_workers=2,
    seed=42,
)

print("=" * 60)
print("V3 AGGRESSIVE CONFIGURATION")
print("=" * 60)
print(f"Epochs: {training_args.num_train_epochs} (V2: 20)")
print(f"Learning rate: {training_args.learning_rate} (V2: 1e-4)")
print(f"Warmup: {training_args.warmup_ratio} (V2: 0.05)")
print(f"LR schedule: linear (V2: cosine)")
print(f"Effective batch: 64 (V2: 32)")
print("=" * 60)
```

```python
# Cell 10: Metrics
VOCAB_SIZE = len(tokenizer)

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    
    predictions = np.clip(predictions, 0, VOCAB_SIZE - 1)
    
    try:
        pred_texts = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    except:
        return {"exact_match": 0.0, "normalized_match": 0.0}
    
    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    labels = np.clip(labels, 0, VOCAB_SIZE - 1)
    
    try:
        label_texts = tokenizer.batch_decode(labels, skip_special_tokens=True)
    except:
        return {"exact_match": 0.0, "normalized_match": 0.0}
    
    exact_matches = 0
    normalized_matches = 0
    total = len(pred_texts)
    
    for pred, label in zip(pred_texts, label_texts):
        if pred.strip() == label.strip():
            exact_matches += 1
        
        pred_norm = normalize_sql(pred)
        label_norm = normalize_sql(label)
        if pred_norm == label_norm:
            normalized_matches += 1
    
    return {
        "exact_match": exact_matches / total if total > 0 else 0.0,
        "normalized_match": normalized_matches / total if total > 0 else 0.0
    }

print("Metrics function ready.")
```

```python
# Cell 11: Initialize Trainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized["train"],
    eval_dataset=tokenized["validation"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

print("Trainer ready (V3 Aggressive).")
print(f"Train: {len(tokenized['train'])} | Eval: {len(tokenized['validation'])}")
```

```python
# Cell 12: Verification
print("Verifying setup...")

test_batch = [tokenized['train'][i] for i in range(2)]
collated = data_collator(test_batch)
print(f"✓ Collator OK")

model.eval()
with torch.no_grad():
    out = model(**{k: v.to(model.device) for k, v in collated.items()})
print(f"✓ Forward OK (loss: {out.loss.item():.2f})")

fake_pred = np.random.randint(0, VOCAB_SIZE, (2, 10))
fake_label = np.random.randint(0, VOCAB_SIZE, (2, 10))
m = compute_metrics((fake_pred, fake_label))
print(f"✓ Metrics OK")

print("\n" + "=" * 60)
print("ALL CHECKS PASSED - V3 READY")
print("=" * 60)
```

```python
# Cell 13: TRAIN
print("=" * 60)
print("STARTING V3 AGGRESSIVE TRAINING")
print("=" * 60)
print(f"Model: {MODEL_NAME}")
print(f"Epochs: 15 | LR: 3e-4 | Effective Batch: 64")
print(f"Expected time: ~1.5-2.5 hours on P100/T4")
print("=" * 60)
print("\nYou can close the browser. Training will continue.\n")

if torch.cuda.is_available():
    torch.cuda.empty_cache()

result = trainer.train()

print("\n" + "=" * 60)
print("V3 TRAINING COMPLETE!")
print("=" * 60)
print(f"Train loss: {result.training_loss:.4f}")
print(f"Time: {result.metrics['train_runtime']/3600:.2f} hours")
```

```python
# Cell 14: Evaluate
print("Evaluating...\n")

eval_results = trainer.evaluate()

print("=" * 60)
print("V3 RESULTS")
print("=" * 60)
print(f"Eval Loss: {eval_results['eval_loss']:.4f}")
print(f"Exact Match: {eval_results['eval_exact_match']*100:.2f}%")
print(f"Normalized Match: {eval_results['eval_normalized_match']*100:.2f}%")
print("=" * 60)

nm = eval_results['eval_normalized_match'] * 100
if nm >= 40:
    print("\nGrade: EXCELLENT")
elif nm >= 30:
    print("\nGrade: GOOD")
elif nm >= 20:
    print("\nGrade: ACCEPTABLE")
else:
    print("\nGrade: NEEDS MORE TRAINING")
```

```python
# Cell 15: Save Model
OUTPUT = "./t2sql_final_v3"

print(f"Saving to {OUTPUT}...")
trainer.save_model(OUTPUT)
tokenizer.save_pretrained(OUTPUT)

report = {
    "version": "v3_aggressive",
    "model": MODEL_NAME,
    "epochs": 15,
    "learning_rate": "3e-4",
    "lr_schedule": "linear",
    "effective_batch_size": 64,
    "train_loss": result.training_loss,
    "eval_loss": eval_results['eval_loss'],
    "exact_match_pct": eval_results['eval_exact_match'] * 100,
    "normalized_match_pct": eval_results['eval_normalized_match'] * 100,
    "training_hours": result.metrics['train_runtime'] / 3600
}

with open("report_v3.json", "w") as f:
    json.dump(report, f, indent=2)

print("Saved!")
```

```python
# Cell 16: Test Predictions
from transformers import pipeline

print("Testing V3 model...\n")

gen = pipeline(
    "text2text-generation",
    model=OUTPUT,
    device=0 if torch.cuda.is_available() else -1
)

def predict_v3(q, s):
    inp = f"Question: {q} Schema: TABLE {s} SQL:"
    out = gen(inp, max_length=256, num_beams=4)
    return out[0]['generated_text']

tests = [
    ("How many students are there?", "students ( id number, name text, age number, gpa number )"),
    ("Find students with GPA above 3.5", "students ( id number, name text, gpa number )"),
    ("List all departments and their average salary", "employees ( id number, name text, dept text, salary number )"),
    ("Show the names of courses with more than 100 students", "courses ( id number, name text, student_count number )"),
]

for q, s in tests:
    sql = predict_v3(q, s)
    print(f"Q: {q}")
    print(f"SQL: {sql}")
    print()
```
